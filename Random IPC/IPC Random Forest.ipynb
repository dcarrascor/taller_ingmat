```python
#Hacer datos legibles

import csv
import numpy
import codecs

matrix=[]
f  = open('datos.csv', "rt",encoding="utf-8")
read = csv.reader(f)
for row in read :
    matrix.append(row)
matrix[0][0]="DIA"
for i in range(1,len(matrix)):
    for j in range(0,3):
        if j==0:
            matrix[i][j]="D"+str(matrix[i][j])
        elif j==1:
            matrix[i][j]="M"+str(matrix[i][j])
        else:
            matrix[i][j]="Y"+str(matrix[i][j])
with open('datos_legibles.csv', 'w', newline='') as csvFile:
    writer = csv.writer(csvFile)
    writer.writerow(["PRODUCTO", "DIA", "MES", "YEAR", "VALOR"])
    for j in range(1, 304):
        nombre=matrix[0][2+j]
        for i in range (1,len(matrix)):
            dia=matrix[i][0]
            mes=matrix[i][1]
            year=matrix[i][2]
            valor=matrix[i][j+2]
            fila=[nombre, dia, mes, year, valor]
            writer.writerow(fila)
csvFile.close()


TODOS_PRODUCTOS=[]
for j in range(1,304):
    TODOS_PRODUCTOS.append(matrix[0][2+j])
```

```python
# Pandas is used for data manipulation
import pandas as pd
import numpy as np
# Read in data and display first 5 rows
features = pd.read_csv('datos_legibles.csv', encoding = "ISO-8859-1", engine='python')
features.head(5)
```

```python
print('The shape of our features is:', features.shape)
```

```python
# One-hot encode the data using pandas get_dummies
features = pd.get_dummies(features)
# Display the first 5 rows of the last 12 columns
temp=np.array(features)
features.head(5)
```

```python
predecir=features.loc[features['VALOR'] == 0]
features=features.loc[features['VALOR'] != 0]
```

```python
# Use numpy to convert to arrays
import numpy as np
# Labels are the values we want to predict
labels = np.array(features["VALOR"])
# Remove the labels from the features
# axis 1 refers to the columns
features= features.drop("VALOR", axis = 1)
predecir= predecir.drop("VALOR", axis = 1)
# Saving feature names for later use
feature_list = list(features.columns)
# Convert to numpy array
features = np.array(features)
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(train_features, train_labels);
```
