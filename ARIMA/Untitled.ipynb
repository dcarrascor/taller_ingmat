```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


#load the data
data = pd.read_csv('input.csv')

#divide into train and validation set
trains=[]
valids=[]
for i in range(50,len(data)-1):
    train = data[:i]
    train=train[['MONTH', "IPC_REAL"]]
    train.drop('MONTH',axis=1,inplace=True)
    trains.append(train)
    valid = data[i:i+1]
    valid=valid[['MONTH', "IPC_REAL"]]
    valid.drop('MONTH',axis=1,inplace=True)
    valids.append(valid)
    train['IPC_REAL'].plot()
    valid['IPC_REAL'].plot()

```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
#building the model
from pmdarima.arima import auto_arima
errr_m=0
conta=0
for i in range(50,len(data)-1):
    train=trains[i-50]
    valid=valids[0]
    model = auto_arima(pd.to_datetime(train.stack()).unstack(), trace=True, error_action='ignore', suppress_warnings=True)
    model.fit(train)

    forecast = model.predict(n_periods=len(valid))
    forecast = pd.DataFrame(forecast,index = valid.index,columns=['Prediction'])

    #plot the predictions for validation set
    plt.plot(train, label='Train')
    plt.plot(valid, label='Valid')
    plt.plot(forecast, label='Prediction')
    plt.show()
    #calculate rmse

    rms = mean_squared_error(valid,forecast)
    errr_m=errr_m+rms
    conta=conta+1
    print(rms)
    try:
        valid.pop(0)
    except:
        a=0
```
